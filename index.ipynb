```python
# escoge solo las columnas de interés
keep=[1,2,4,6,8,10,12,14,16]
dontKeep=set(data.columns.to_list())-set(data.columns[keep])
dataF=data.drop(columns=dontKeep)

dataF
```

```python
data_frame=dataF[9:203]
data_frame
```

```python
# asignar nombres a las columnas
change1={'Unnamed: 1':"Num"}
data_frame.rename(columns=change1,inplace=True)
data_frame
change2={'Unnamed: 2':"Country"}
data_frame.rename(columns=change2,inplace=True)
data_frame
change3={'Unnamed: 4':"Human Development Index"}
data_frame.rename(columns=change3,inplace=True)
change4={'Unnamed: 6':"Life expectancy at birth"}
data_frame.rename(columns=change4,inplace=True)
change5={'Unnamed: 8':"Expected years of schooling"}
data_frame.rename(columns=change5,inplace=True)
change6={'Unnamed: 10':"Mean years of schooling"}
data_frame.rename(columns=change6,inplace=True)
change7={'Unnamed: 12':"Gross national income per capita"}
data_frame.rename(columns=change7,inplace=True)
change8={'Unnamed: 14':"GNI per capita rank minus HDI rank"}
data_frame.rename(columns=change8,inplace=True)
change9={'Unnamed: 16':"HDI RANK"}
data_frame.rename(columns=change9,inplace=True)
data_frame
```

```python
data_frame.info()
```

```python
# limpiar nombre de columnas, sin epacios
data_frame.columns=data_frame.columns.str.replace('\s','_',regex=True)
data_frame
```

```python

data_frame[data_frame.Expected_years_of_schooling.str.contains(r'\D')]
data_frame[data_frame.Life_expectancy_at_birth.str.contains(r'\D')]
data_frame[data_frame.HDI_Rank.str.contains(r'\D')]
data_frame[data_frame.Mean_years_of_schooling.str.contains(r'\D')]
data_frame[data_frame.Life_expectancy_at_birth.str.contains(r'\D')]
data_frame[data_frame.GNI_per_capita_rank_minus_HDI_rank(r'\D')]
```
